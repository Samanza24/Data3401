## How do we select multiple rows and columns in pandas dataframe?

```python
# loc and iloc
```

### The loc method is used to select rows and columns by label.

#### A single label, a list of labels, a slice labels, a boolen Series, A colon(which indicates all label)

```python
# row 0, all columns
df_listing.loc[0, :]
```

```python
# row 0, 2, 4, all columns 

df_listing.loc[[0, 2, 4], :]
```

```python
# rows 0 through 2(inclusive), all columns

df_listing.loc[0:2, :]
```

```python
# rows 0 through 2(inclusive), column host_name
df_listing.loc[0:2 , 'host_name']
```

```python
# rows 0 through 2(inclusive), column host_name and host_id
df_listing.loc[0:2, ['host_id', 'host_name']]
```

```python
df_listing[0:3][['host_id', 'host_name']]
```

```python
df_listing[['host_id', 'host_name']].head(3)
```

```python
# rows 0 through 2(inclusive), column host_id through Price(inclusive)
df_listing.loc[0:2, 'host_id' : 'price']
```

```python
# all rows which the host_name is Karen and room_type

df_listing.loc[df_listing['host_name'] == 'Karen', 'room_type']
```

```python
df_listing[df_listing['host_name'] == 'Karen']['room_type']
```

## iloc

### The iloc method is used to select rows and columns by integer position. You can do
### a single integer position, a list of integer position, a slice integer position and a colon (indicates all integer positions)

```python
# rows in positions 2 and 4, columns in position 2 and 4
df_listing.iloc[[2, 4], [2, 4]]
```

```python
df_listing.info()
```

```python
# rows in position 0 through 2(exclusive) and columns in position 0 through 5(exclusive)

df_listing.iloc[0:2, 0:5]
```

## Apply Function

```python
df_titanic = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df_titanic.head()
```

```python
df_titanic['Embarked'].unique()
```

```python
## map S >> 0 , C >> 1 and Q >>2
# MAP METHOD

df_titanic['Embarked_Num'] = df_titanic['Embarked'].map({'S' : 0,
                                                         'C' : 1,
                                                         'Q' : 2})
```

```python
df_titanic.sample(10)
```

```python
# Goal : Apply a function to each element in a Series
# Method :  apply
# Note : map method can be substituted for apply in many cases, but apply is more flexible
# so it is recommended
```

```python
# apply the 'max' function aliong with axis=0 to calculate the maximum of each column 
```

```python
df_titanic.loc[:, ['Pclass', 'Fare']].apply(max, axis = 0)
```

```python
# applymap (dataframe method)
```

```python
df_titanic.loc[:, ['Pclass', 'Fare']].applymap(int).head()
```

# DataFrame Merge / Join / Append

```python
df1 = pd.DataFrame({'color' : ['green', 'yellow', 'red'],
                    'num' : [1, 2, 3]})
df2 = pd.DataFrame({'color' : ['green', 'yellow', 'pink'],
                    'size' : ['S', 'M', 'L']})
```

```python
df1
```

```python
df2
```

```python
# Inner Join
```

```python
pd.merge(df1,
         df2,
         how = 'inner',
         on = 'color')
```

```python
# Left Join - Include all observations found in df1
pd.merge(df1,
         df2,
         how = 'left',
         on = 'color')
```

```python
# Right Join - Include all observations found in df2
pd.merge(df1,
         df2,
         how = 'right',
         on = 'color')
```

```python
# Outer Join - Include all observations found in either df1 or df2
pd.merge(df1,
         df2,
         how = 'outer',
         on = 'color')
```

```python
df1 = df1.rename(columns = {'color' : 'clr'})
```

```python
df1
```

```python
df2
```

```python
# Left Join - Include all observations found in df1
pd.merge(df1,
         df2,
         how = 'left',
         left_on = 'clr',
         right_on = 'color')
```
